```python
import pandas as pd

df = pd.read_csv(
    "tablas/ajustes_inventario_registro.csv"
)

if "Día" not in df.columns:
    df.insert(1, "Día", "")

if "Mes" not in df.columns:
    df.insert(2, "Mes", "")

if "Año" not in df.columns:
    df.insert(3, "Año", "")

df.to_csv(
    "tablas/ajustes_inventario_registro.csv",
    index=False
)

print("Columnas agregadas correctamente.")
print(df.columns.tolist())
```
