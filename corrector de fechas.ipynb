```python
import pandas as pd

df_ventas = pd.read_csv(
    "tablas/ventas_df.csv",
    sep=";"
)

dias_semana = [
    "lunes",
    "martes",
    "miércoles",
    "jueves",
    "viernes",
    "sábado",
    "domingo"
]

meses_texto = [
    "",
    "enero",
    "febrero",
    "marzo",
    "abril",
    "mayo",
    "junio",
    "julio",
    "agosto",
    "septiembre",
    "octubre",
    "noviembre",
    "diciembre"
]

def convertir_fecha(fecha):

    try:

        fecha_dt = pd.to_datetime(
            fecha,
            dayfirst=True
        )

        return (
            f"{dias_semana[fecha_dt.weekday()]}, "
            f"{fecha_dt.day} de "
            f"{meses_texto[fecha_dt.month]} de "
            f"{fecha_dt.year}"
        )

    except:

        return fecha

df_ventas["fecha"] = (
    df_ventas["fecha"]
    .astype(str)
    .apply(convertir_fecha)
)

df_ventas.to_csv(
    "tablas/ventas_df.csv",
    sep=";",
    index=False
)

print("Fechas corregidas")
```
